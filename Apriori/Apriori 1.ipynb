```python
# import the libraries

import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
# read data

df = pd.read_csv('../Data/GroceryStoreDataSet.csv',names=['products'],header=None)
```

```python
data = list(df["products"].apply(lambda x:x.split(',')))
data 
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
df1 = apriori(df,min_support=0.1,use_colnames=True , max_len=4)
df1.sort_values(by="support",ascending=False)
```

```python
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1.sort_values(by="support",ascending=False)
```

```python
df1[(df1['length']==2) & (df1['support']>=0.1)].sort_values(by="support",ascending=False)
```
